[] function arguments and parameters
- in a positional argument, user specifies the value of the argument only. order matters
- in a keyword arugment, user specifies the parameter name and its value. order does not matter
- / specifies that every argument before it must be positional
- * followed by a comma specifies that every argument after it is a keyword arg
- *args stores a practically limitless number of inputs into a tuple called args
- **kwargs stores a practicall limitless number of keyword arguments in a dict
